## Objective: Visualize whether Sensor A & B in each PurpleAir Unit agree with one another
#### Specify directory
#### Associate sensors with the months they were collocated
#### Parameters visualizing between A&B
#### Loop through folder of sensors 
#### Plot A vs B in a scatterplot



```python
import pandas as pd                 # pandas allows me to read and manipulate my csv data with python in dataframes
from datetime import datetime       # datetime allows me to parse through datetime timestamps in cells and isolate specific parts of the timestamp (e.g. hour, minute sec)
import os                           # os allows me to run the code on each file within a folder
import scipy                        # scipy calculates statistical measurements needed
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

computer = 'sph0088'

#Specify Directory
directory = os.listdir(r"C:\Users\{}\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II_PA-II-SD".format(computer))
os.chdir(r"C:\Users\{}\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II_PA-II-SD".format(computer))

#Associate sensor with the collocation month
April = ['PA-80f1', 'PA-80f8','PA-8038','PA-8095','PA-a5f0']
June = ['PA-8d18', 'PA-803e','PA-8079','PA-adb9','PA-c074']
July = ['PA-16a0-NoSD','PA-38da-NoSD','PA-d5ec-NoSD','PA-d147-NoSD']
August = ['PA-82c0', 'PA-198','PA-8060','PA-e3ba','PA-e4bd','PA-e4de', 'PA-e4f6','PA-e396']

#Parameters visualizing between A&B
x='pm2_5_atm'
y='pm2_5_atm_b'

#Loop through folder of sensors and plot A vs B in a scatterplot
for file in directory:
    df = pd.read_csv(file)
#     df = df.loc[(df['UTCDateTime'] > '2020/03/31') & (df['UTCDateTime'] <= '2020/10/01')] #filter dataframe by date
#     df['UTCDateTime'] = pd.to_datetime(df['UTCDateTime'], format="%Y/%m/%dT%H:%M:%Sz") #convert to datetime
#     df['UTCDateTime'] = df['UTCDateTime'].dt.floor('H') #Drops off the minutes and seconds of the time
#     df = pd.DataFrame(df.groupby(['UTCDateTime','mac_address'],as_index=False).mean()) #groupby hour
    filename = os.path.splitext(file)[0] #split extension from filename
#     print(filename)
#     df.to_csv(r'C:\Users\sph0088\OneDrive - UNT System\AQ\Calibration_Sensors\Collocation_Data\PA-II_PA-II-SD\{}.csv'.format(filename),index=False)
    if filename in April:
        df = df[(df['UTCDateTime'] > '2020-04-11')]
    elif filename in June:
        df = df[(df['UTCDateTime'] > '2020-05-31')]
    elif filename in July:
        df = df[(df['UTCDateTime'] > '2020-06-27')]
    elif filename in August:
        df = df[(df['UTCDateTime'] > '2020-08-04')] 
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(df[x], df[y])
    print(str(filename) + ' = ' + str(r_value**2))
    plt.plot(df[x], df[y], 'o', label='original data')#,title='R2')
    plt.plot(df[x], intercept + slope*df[x], 'r',label='fitted line')#+ str(r_value**2))
    # plt.title('R2:' + str(r_value**2))
    plt.title(str(filename)+' R2: %.3f'% + r_value**2)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
    plt.show()


```
